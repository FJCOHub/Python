```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# Cargar datos
df = pd.read_csv("../data/ventas.csv")

# Gráfico de ventas por tienda
plt.figure(figsize=(6,4))
sns.barplot(x="Tienda", y="Ventas", data=df.groupby("Tienda").sum().reset_index(), palette="Blues_d")
plt.title("Ventas Totales por Tienda")
plt.savefig("../images/ventas_tienda.png")  # Guardar imagen
plt.show()

# Gráfico de ventas por mes y tienda
plt.figure(figsize=(6,4))
sns.lineplot(x="Mes", y="Ventas", hue="Tienda", data=df, marker="o")
plt.title("Ventas por Mes y Tienda")
plt.savefig("../images/ventas_mes.png")  # Guardar imagen
plt.show()
```
